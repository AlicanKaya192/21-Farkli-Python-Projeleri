# 🎥 Video Veri Boyutu Hesaplama Uygulaması

Bu proje, seçilen bir video çözünürlüğüne (360p, 480p, 720p, 1080p) ve girilen video süresine (dakika) bağlı olarak tahmini video dosya boyutunu (MB) hesaplayan basit bir **Tkinter GUI** uygulamasıdır.

Bu notebook, uygulamanın arka planında dönen matematiksel formülleri, kütüphane kullanımlarını ve arayüz elemanlarının yapısını adım adım açıklamaktadır.

## 1. Adım: Gerekli Kütüphanelerin İçe Aktarılması

```python
import tkinter as tk
from tkinter import ttk, messagebox
```

## 2. Adım: Çözünürlük Standartları ve Bit Hızları (Bitrates)

Video dosya boyutu, doğrudan videonun saniyede taşıdığı veri miktarı olan **Bit Hızı (Bitrate)** ile ilgilidir. Genel platform standartları baz alınarak çözünürlüklere göre bit hızları (kbps - kilobits per second) şu şekilde belirlenmiştir:

- **360p:** 750 kbps
- **480p:** 1500 kbps (1.5 Mbps)
- **720p:** 3000 kbps (3.0 Mbps)
- **1080p:** 4500 kbps (4.5 Mbps)

```python
# Bit hızlarının tanımlanması
bit_rates = {
    "360p": 750, 
    "480p": 1500, 
    "720p": 3000,
    "1080p": 4500
}
```

## 3. Adım: Boyut Hesaplama Formülü

Videonun dosya boyutu şu formülle hesaplanır:

$$\text{Dosya Boyutu (MB)} = \frac{\text{Bit Hızı (kbps)} \times \text{Süre (saniye)}}{8 \times 1024}$$

Süre arayüzden **dakika** olarak alındığı için saniyeye çevirmek amacıyla 60 ile çarpılır:

$$\text{Dosya Boyutu (MB)} = \frac{\text{Bit Hızı} \times (\text{Dakika} \times 60)}{8 \times 1024}$$

Burada:
- `8`'e bölme işlemi bit değerini **Byte**'a çevirir.
- `1024`'e bölme işlemi kilobyte değerini **Megabyte (MB)** değerine çevirir.

```python
# Örnek Hesaplama:
# 1080p çözünürlükte 10 dakikalık bir videonun boyutu:
resolution = "1080p"
duration_minutes = 10

bit_rate = bit_rates[resolution]
duration_seconds = duration_minutes * 60

# Hesaplama formülü
size_mb = (bit_rate * duration_seconds) / 8 / 1024
print(f"{resolution} çözünürlüğünde {duration_minutes} dakikalık video: {size_mb:.2f} MB")
```

## 4. Adım: Arayüz Fonksiyonu (hesapla)

Kullanıcı 'Hesapla' butonuna tıkladığında çalışan fonksiyondur. Arayüz elemanlarından veriyi çeker, geçerli bir tamsayı olup olmadığını kontrol eder (`try-except`) ve sonucu arayüze basar.

```python
def hesapla_test(resolution, duration_str):
    try:
        # Arayüzden alınan string girdiyi tamsayıya çevir
        duration = int(duration_str)
        if duration <= 0:
            raise ValueError("Süre sıfır veya negatif olamaz.")
            
        bit_rate = bit_rates[resolution]
        file_size_mb = bit_rate * duration * 60 / 8 / 1024
        return f"Video Boyutu: {file_size_mb:.2f} MB"
    except ValueError:
        return "Hata: Lütfen geçerli bir süre (pozitif tamsayı) giriniz."

# Testler
print(hesapla_test("720p", "15"))
print(hesapla_test("1080p", "abc")) # Hatalı girdi testi
```

## 5. Adım: Arayüz Elemanlarının Oluşturulması ve Konumlandırma

Tkinter penceremizde kullanılan elemanlar ve görevleri:

1. **`root` Penceresi:** `root = tk.Tk()` ile oluşturulur. `resizable(False, False)` ile pencerenin genişletilmesi engellenir.
2. **`ttk.Style`:** Bileşenlerin font ve renk ayarlarını yönetir. Butonun aktif/pasif hover durumları `style.map` ile ayarlanır.
3. **`ttk.Combobox`:** Çözünürlük listesini (`360p`, `480p`, `720p`, `1080p`) gösteren salt okunur (`state="readonly"`) seçim kutusudur.
4. **`ttk.Entry`:** Kullanıcının video süresini dakika bazında girdiği metin giriş alanıdır.
5. **`ttk.Button`:** Hesaplama tetikleyicisi butondur. Tıklandığında `hesapla` fonksiyonunu çalıştırır.
6. **`result_label`:** Sonucu ekranda gösteren metin etiketidir.
