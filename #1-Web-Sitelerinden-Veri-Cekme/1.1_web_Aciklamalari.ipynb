# 🌐 Temel Web Scraping Kullanımı

Bu proje, `BeautifulSoup` kütüphanesinin en temel kullanım yöntemlerini ve etiket (HTML tag) seçicilerini gösteren bir pratik dosyasıdır.

Aşağıdaki kod hücrelerini sırasıyla çalıştırarak web scraping işleminin temel taşlarını öğrenebilirsiniz.

## 1. Adım: Siteye Bağlanma ve Durum Kontrolü

Bir siteye veri çekmek için gitmeden önce 'kapıyı çalmamız' ve izin almamız gerekir. `requests` kütüphanesi bu işe yarar.

```
import requests
from bs4 import BeautifulSoup

# Alican Kaya'nın kişisel sitesine bağlanma isteği gönderiyoruz
url = "https://alican-kaya.com/"
response = requests.get(url)

# Sunucudan dönen yanıt 200 ise işlem başarılıdır
if response.status_code == 200:
    print("Siteden Veri Çekilebilir")
else:
    print(f"Siteden Veri Çekilemez. Hata Kodu: {response.status_code}")
```

## 2. Adım: HTML Kodlarını Okunabilir Hale Getirme (Parsing)

Siteden gelen karmaşık HTML yığınını, Python'un içinde rahatça dolaşabileceği mantıklı bir ağaç yapısına çevirmek için `BeautifulSoup` kullanırız.

```
# 'html.parser' parametresi ile HTML'i ayrıştır
soup = BeautifulSoup(response.content, "html.parser")

print("Sayfanın Sekme Başlığı:")
print(soup.title) # Sadece <title> etiketini yazdırır

print("\nHTML Kodunun Düzenli Hali (prettify):")
# prettify() fonksiyonu HTML kodunu girintili ve estetik (okunaklı) bir formata sokar
print(soup.prettify()[:500]) # Sadece ilk 500 karakteri yazdıralım, çok uzun olmasın
```

## 3. Adım: Belirli Etiketleri ve Sınıfları (Class) Seçme

Web sayfasındaki belirli bir yazıya veya resme ulaşmak için, onun `div`, `li`, `span` gibi etiketlerini ve onlara ait özel `class` (sınıf) isimlerini hedef alırız.

```
print("--- BELİRLİ BİR SINIFI SEÇME ---")
try:
    # find() metodu sayfada bu özelliğe uyan İLK elementi bulup getirir
    div_cek = soup.find("div", {"class": "h-28"})
    
    # Sadece içindeki yazıyı (.text) alıyoruz
    print(div_cek.text)
    
    # Eğer bu div'in içinde bir link olsaydı .get('href') ile o linki çekebilirdik
    print(div_cek.get("href")) 
except AttributeError:
    # Web sitesinin tasarımı değiştiyse ve o sınıf artık yoksa programın çökmesini engelleriz
    print("Sınıfı 'h-28' olan bir <div> elementi bulunamadı.")

print("\n--- BAŞKA BİR LİSTE ELEMANI SEÇME ---")
try:
    # Sınıfı belirterek liste elemanını (li) çek
    li_cek = soup.find("li", {"class": "flex items-start gap-2 text-sm text-white/90"}).text
    print(li_cek)
except AttributeError:
    print("Belirtilen sınıfa ait bir <li> elementi bulunamadı.")
```
