# 🦠 Korona Verisi Çekme (Tablo Kazıma)

Bu proje, `covid19.saglik.gov.tr` (veya benzeri bir veri tablosu içeren) sitesinden, HTML içindeki tablo verilerini otomatik olarak okumayı ve ayıklamayı öğretir.

## Adım Adım Tablo Okuma İşlemi

Tablolar HTML'de `<table>` (Tablo), `<tbody>` (Gövde) ve `<tr>` (Satır) etiketleri ile kurulur. Biz de verileri çekerken bu etiket hiyerarşisine göre arama yapacağız.

```
import requests
from bs4 import BeautifulSoup

# Siteye bağlan
url = "https://covid19.saglik.gov.tr/"
response = requests.get(url)

if response.status_code == 200:
    print("Siteden Veri Çekilebilir\n")
else:
    print(f"Siteden Veri Çekilemez. Hata Kodu: {response.status_code}")

# HTML'i ayrıştır
soup = BeautifulSoup(response.content, "html.parser")

try:
    # Önce sayfanın neresinde olursa olsun 'tbody' (Tablo gövdesi) etiketini bul.
    # Ardından o gövdenin içindeki TÜM 'tr' (Tablo satırı) etiketlerini bir liste olarak al.
    tablo_satirlari = soup.find("tbody").find_all("tr")
    
    # Tüm satırları döngüyle tek tek gez
    for satir in tablo_satirlari:
        print("----------")
        # O satırın içindeki tüm hücreleri saf metin olarak alıp ekrana yazdır
        print(satir.text)
        
except AttributeError:
    # Sağlık Bakanlığı sitesini sık sık güncellediğinden dolayı 
    # tablo bazen ana sayfada bulunmayabilir, bu yüzden hata yakalama koyuyoruz.
    print("Sitede 'tbody' etiketi bulunamadı! Sayfa yapısı değişmiş veya veriler JavaScript ile sonradan yükleniyor olabilir.")
```

# 🔄 Tablo Çekme Mantığı ve Akış

### 1. Kapsayıcıyı Bulmak
Tıpkı bir soğanı soymak gibi, içteki verilere ulaşmak için önce dış kabuğu bulmamız gerekir. Bu yüzden doğrudan verilere saldırmaz, önce `.find("tbody")` diyerek ana tablo gövdesini alırız.

### 2. Listeyi Çıkartmak
Ana gövdeyi bulduktan sonra, `.find_all("tr")` komutunu kullanarak o tablonun içindeki 1. satır, 2. satır, 3. satır gibi tüm elemanları bir Python **listesi (array)** haline getiririz.

### 3. Döngüyle Ayıklamak
Oluşan bu satırlar listesini bir `for` döngüsüne sokarız. Her adımda döngü bir sonraki satırı eline alır, `.text` metodu sayesinde o satırın içindeki çirkin HTML kodlarını çöpe atar ve geriye sadece insanların okuyabileceği temiz veriyi bırakarak ekrana yazar.
