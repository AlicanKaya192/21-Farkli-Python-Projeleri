# ⏰ Dijital Masaüstü Saati Uygulaması

Bu proje, bilgisayarınızda anlık olarak çalışan, saati ve tarihi canlı olarak saniye saniye gösteren şık ve siyah temalı bir **Masaüstü Dijital Saati** uygulamasıdır.

Aşağıdaki hücreleri **sırasıyla (Shift + Enter)** çalıştırarak arayüzün nasıl çizildiğini ve saatin nasıl saniye saniye güncellendiğini öğrenebilirsiniz.

## 1. Adım: Kütüphanelerin İçe Aktarılması

Görsel bir pencere oluşturmak için `tkinter`, sistemin anlık saat ve tarih bilgisini almak için ise Python'un yerleşik `time` kütüphanesini kullanacağız.

```
from tkinter import Label, Tk
import time
```

## 2. Adım: Saati ve Tarihi Güncelleyen Motor (Fonksiyon)

Uygulamanın kalbi burasıdır. Bu fonksiyon sistem saatini alır, metne çevirir, ekrandaki yazıyı değiştirir ve **kendi kendini tekrar çağırarak** saatin akmasını sağlar.

```
def digital_clock():
    # Sistemden anlık saati 'Saat:Dakika:Saniye' formatında al
    time_live = time.strftime("%H:%M:%S")
    # Saat etiketindeki yazıyı bu yeni saat ile değiştir
    label.config(text=time_live)

    # Sistemden tarihi 'Gün Ay Yıl' formatında al
    date_info = time.strftime("%d %B %Y")
    # Tarih etiketindeki yazıyı bu yeni tarih ile değiştir
    date_label.config(text=date_info)

    # EN ÖNEMLİ SATIR: 
    # 200 milisaniye (0.2 saniye) sonra bu fonksiyonu ('digital_clock') YENİDEN ÇALIŞTIR!
    label.after(200, digital_clock)
```

## 3. Adım: Ana Pencere ve Stil Ayarları

Fonksiyonumuz hazır olduğuna göre, artık saati göstereceğimiz o siyah, şık pencereyi ve yazılarda kullanacağımız renk ve fontları ayarlıyoruz.

```
# Ana pencereyi oluştur
app_window = Tk()
app_window.title("Dijital Saat")
app_window.geometry("350x300") # Genişlik x Yükseklik
app_window.resizable(False, False) # Pencerenin fareyle büyütülüp küçültülmesini engelle
app_window.configure(bg="black") # Arka planı siyah yap

# Ekranda yazılacak saat ve tarih için ortak tasarım ayarları (değişkenler)
text_font = ("Boulder", 36, 'bold') # Boulder fontu, 36 punto, kalın (bold)
background = "black"
foreground = "White" # Yazı rengi beyaz
border_width = 20
```

## 4. Adım: Etiketlerin Çizilmesi ve Programın Başlatılması

Yukarıdaki tasarımları kullanarak ekrana saat ve tarih için iki ayrı etiket (`Label`) koyuyoruz. Ardından fonksiyonumuzu **ilk kez** çağırıp döngüyü başlatıyoruz.

**NOT:** Aşağıdaki hücreyi çalıştırdığınızda Dijital Saat ekranda belirecektir ve çalışmaya devam edecektir. İşiniz bittiğinde saati çarpıdan kapatmayı unutmayın.

```
# Saat etiketi oluşturuldu ve ekrana (row=0 yani ilk satıra) yerleştirildi
label = Label(app_window, font=text_font, bg=background, fg=foreground, border=border_width)
label.grid(row=0, column=1, padx=10, pady=10)

# Tarih etiketi oluşturuldu ve saat etiketinin tam altına (row=1 yani ikinci satıra) yerleştirildi
date_label = Label(app_window, font=text_font, bg=background, fg=foreground, border=border_width)
date_label.grid(row=1, column=1, padx=10, pady=10)

# MOTORU ÇALIŞTIR! Fonksiyonu ilk kez biz çağırıyoruz, sonrasını o kendisi (after ile) halledecek.
digital_clock()

# Pencerenin kapanana kadar açık kalmasını ve saniye saniye güncellenmesine izin vermesini sağla.
app_window.mainloop()
```

# 🔄 Programın Çalışma Mantığı ve Kodların Gerçek İşleyiş Sırası

Python bu projeyi çalıştırdığında, sadece durağan bir pencere çizmekle kalmaz; arka planda o saatin nasıl tik-tak ederek güncellendiğini anlamak için işleyişi adım adım inceleyelim:

### Aşama 1: Temel Arayüzün ve Ayarların Okunması
Python önce en tepedeki fonksiyonu (`digital_clock`) ezberler ama içine girmez. Doğrudan aşağıya inip ana pencereyi yaratır, boyutlandırır ve saat ile tarihi göstereceği "boş" etiketleri ekrana yapıştırır.

```
app_window = Tk() # 1- Ana çerçeve yaratıldı.
# ... pencere boyutlandırıldı ve siyah renk verildi...

label = Label(...) # 2- İçi henüz boş olan saat etiketi eklendi.
date_label = Label(...) # 3- İçi henüz boş olan tarih etiketi eklendi.
```

### Aşama 2: Motorun İlk Ateşlenmesi
Arayüz çizildikten sonra, Python aşağıya doğru inmeye devam eder ve `digital_clock()` komutuyla karşılaşır. İşte bu satır, saatin ilk saniyesini ekrana vuracak olan tetiği çeker. Python buradan yukarıdaki fonksiyonun içine zıplar.

```
digital_clock() # 4- İlk ateşleme yapıldı! Şimdi fonksiyona zıpla.
```

### Aşama 3: Saati Alma ve Kendini Döngüye Sokma (Sonsuz Döngü Sırrı)
Fonksiyona giren Python, sistemden o anki saati ve tarihi alır, o az önce yarattığı "boş" etiketlerin içine yazıları yerleştirir (`config` ile). Ancak işin sihri son satırdadır. 

`label.after(200, digital_clock)` komutu, Python'a şunu söyler: *"Tamam, işin bitti. Ama 200 milisaniye (saniyenin 5'te 1'i) bekle ve sonra beni gizlice baştan bir daha çalıştır!"*

Böylece fonksiyon kendi kendini bir döngüye (recursion) hapseder.

```
def digital_clock():
    # 5- Sistemden o anın saat ve tarihi alındı
    time_live = time.strftime("%H:%M:%S")
    
    # 6- Ekranda var olan yazılar bu yeni saniyeyle değiştirildi
    label.config(text=time_live)
    date_label.config(text=date_info)

    # 7- SİHİRLİ DÖNGÜ: 200 milisaniye sonra beni GİZLİCE tekrar çağır!
    label.after(200, digital_clock)
```

### Aşama 4: Dinleme Modu (Mainloop)
İlk ateşleme ve döngü emri verildikten sonra Python fonksiyondan çıkar ve en alt satıra, yani `app_window.mainloop()` komutuna ulaşır.

Bu komut programı açık tutar. Arayüz ekranda kalır ve `after` komutunun yarattığı o gizli geri sayım her 200 milisaniyede bir dolarak fonksiyonu tekrar tekrar tetiklemeye devam eder. Böylece saatimiz, arayüz asla donmadan, her an saniye saniye değişerek **canlı** kalır.

```
app_window.mainloop() # 8- Program burada beklerken, arka planda saat saniyede 5 kez güncellenmeye devam eder!
```
