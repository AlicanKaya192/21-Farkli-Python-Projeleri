# 🔗 Link Kısaltma Uygulaması

Bu proje, uzun web bağlantılarını (URL) **TinyURL API** kullanarak saniyeler içinde kısaltan ve tek tıkla panoya kopyalamanızı sağlayan bir masaüstü uygulamasıdır.

Aşağıdaki hücreleri **sırasıyla (Shift + Enter)** çalıştırarak kodun nasıl adım adım inşa edildiğini inceleyebilirsiniz.

## 1. Adım: Gerekli Kütüphanelerin İçe Aktarılması

Projeyi ayağa kaldırmak için bazı temel kütüphanelere ihtiyacımız var:

* **`tkinter`**: Python'un arayüz çizicisidir. Butonları, giriş kutularını ve pencereyi bu kütüphane sayesinde oluştururuz.
* **`requests`**: İnternetteki başka bir sunucuyla (bu projede TinyURL) iletişime geçip ondan veri almamızı veya veri göndermemizi sağlar. Kısacası programımızın internete açılan kapısıdır.
* **`pyperclip`**: İşletim sisteminizin panosuna erişip metin kopyalamayı (sanki klavyeden `Ctrl+C` yapmışsınız gibi) sağlayan pratik bir kütüphanedir.

```
import tkinter as tk
from tkinter import messagebox
import requests
import pyperclip
```

## 2. Adım: Arka Planda Çalışan Fonksiyonlar

Butonlara basıldığında programın ne yapması gerektiğini fonksiyonlarla belirleriz. İki temel işlevimiz var: Kısaltmak ve Kopyalamak.

### 2.1 Linki Kısaltan Fonksiyon (`shorten_url`)

Bu fonksiyon TinyURL'nin yazılımcılara sunduğu ücretsiz API'yi kullanır. İnternet sitesine elle uzun linki yapıştırmak yerine bunu arka planda kodumuz yapar ve cevabı alır.

```
def shorten_url():
    # Kullanıcının giriş kutusuna yazdığı uzun linki alırız
    long_url = entry.get()
    
    try:
        # TinyURL API'sine uzun linki ekleyerek GET isteği (request) göndeririz
        response = requests.get(f'http://tinyurl.com/api-create.php?url={long_url}')
        
        # Gelen cevabın içinden bize lazım olan metni çekeriz
        short_url = response.text
        
        # Çektiğimiz kısa linki arayüzdeki etiket (label) üzerine yazarız
        result_label.config(text=f"Kısa URL Adresi: {short_url}")
        
        # Link kısaltıldığına göre artık 'Kopyala' butonunu tıklanabilir yaparız
        copy_button.config(state=tk.NORMAL)
    except Exception as e:
        # Olası bir internet kesintisinde çökmeyi engellemek için hata mesajı çıkarırız
        messagebox.showerror("Hata", f"Link kısaltılamadı: {e}")
```

### 2.2 Kısa Linki Panoya Kopyalayan Fonksiyon (`copy_to_clipboard`)

Ekranda yazan metinden sadece link kısmını çekip hafızaya kopyalayan basit ama kullanışlı fonksiyondur.

```
def copy_to_clipboard():
    # Ekranda "Kısa URL Adresi: https://tinyurl..." yazıyor
    # İlk 17 karakter "Kısa URL Adresi: " olduğu için [17:] diyerek orayı kesip atarız
    short_url = result_label.cget("text")[17:]
    
    # Kalan saf linki pyperclip ile panoya kopyalarız
    pyperclip.copy(short_url)
    
    # Kullanıcıyı bilgilendiririz
    messagebox.showinfo("Kopyalandı", "Kısa URL Kopyalandı")
```

## 3. Adım: Kullanıcı Arayüzünün (GUI) Kurulması

Penceremizi oluşturduğumuz ve butonları/kutuları içeriye dizdiğimiz yerdir.

**NOT:** Bu hücreyi çalıştırdığınızda bilgisayarınızda bir pencere açılacak ve uygulamanız tamamen çalışır hale gelecektir. İşiniz bitene kadar bu hücre beklemeye geçecektir.

```
# İşlem bütünlüğünü sağlamak için fonksiyonları eksiksiz tanımlıyoruz
def shorten_url():
    long_url = entry.get()
    try:
        response = requests.get(f'http://tinyurl.com/api-create.php?url={long_url}')
        short_url = response.text
        result_label.config(text=f"Kısa URL Adresi: {short_url}")
        copy_button.config(state=tk.NORMAL)
    except Exception as e:
        messagebox.showerror("Hata", f"Link kısaltılırken hata oluştu: {e}")

def copy_to_clipboard():
    short_url = result_label.cget("text")[17:]
    pyperclip.copy(short_url)
    messagebox.showinfo("Kopyalandı", "Kısa URL Kopyalandı")

# ----------------- ARAYÜZ KODLARI -----------------
app = tk.Tk()
app.title("Link Kısaltma")

label = tk.Label(app, text="Uzun Linki Giriniz: ")
label.pack(pady=10)

entry = tk.Entry(app, width=40)
entry.pack()

shorten_button = tk.Button(app, text="Kısalt", command=shorten_url)
shorten_button.pack()

result_label = tk.Label(app, text="")
result_label.pack()

# Başlangıçta kopyalanacak bir şey olmadığı için state=tk.DISABLED yapılarak buton pasif tutulur
copy_button = tk.Button(app, text="Kopyala", command=copy_to_clipboard, state=tk.DISABLED)
copy_button.pack()

app.mainloop()
```
