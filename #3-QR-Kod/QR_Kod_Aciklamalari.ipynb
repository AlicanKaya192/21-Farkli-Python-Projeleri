# 🔲 QR Kod Oluşturucu Uygulaması

Bu proje, kullanıcının girdiği herhangi bir metni veya URL (bağlantı) adresini alarak bir **QR Kod (Karekod)** oluşturan ve bunu bilgisayara `.svg` (vektörel görsel) formatında kaydeden bir masaüstü uygulamasıdır.

Aşağıdaki hücreleri **sırasıyla (Shift + Enter)** çalıştırarak kodun nasıl adım adım inşa edildiğini inceleyebilirsiniz.

## 1. Adım: Gerekli Kütüphanelerin İçe Aktarılması

QR kod üretmek ve arayüz çizmek için üç temel araca ihtiyacımız var:
* **`tkinter`**: Python'un standart görsel arayüz (GUI) oluşturucusudur. Pencereler ve butonlar için kullanırız.
* **`filedialog`**: Çıkan QR kodunu nereye kaydedeceğimizi soran o standart Windows 'Farklı Kaydet' penceresini açmamızı sağlar.
* **`pyqrcode`**: Verilen metni veya URL'yi hızlıca QR koda dönüştüren ana kütüphanedir.

```
import tkinter as tk
from tkinter import filedialog
import pyqrcode
```

## 2. Adım: QR Kod Oluşturma Fonksiyonu

Bu fonksiyon, uygulamanın 'beynidir'. Butona tıklandığında çalışarak arayüzdeki yazıyı alır, QR koda çevirir ve kaydeder.

```
def qr_kodu_olustur():
    # Arayüzdeki (url_girdi) kutusunun içine yazılan metni/linki alır
    url = url_girdi.get()
    
    if url: # Eğer kutu boş değilse işleme başlar
        # Metni arka planda bir QR Kod objesine dönüştürür
        qr_url = pyqrcode.create(url)
        
        # Kullanıcıya dosyayı nereye kaydedeceğini sorar (Farklı Kaydet penceresi)
        dosya_yolu = filedialog.asksaveasfilename(defaultextension=".svg", filetypes=[("SVG dosyaları", "*.svg")])

        if dosya_yolu: # Eğer kullanıcı kaydetmeyi iptal etmediyse
            # QR kodu belirtilen yere kaydeder. (scale=8 karekodun piksel büyüklüğüdür)
            qr_url.svg(dosya_yolu, scale=8)
            
            # İşlemin bittiğini arayüzdeki etikete yazdırarak haber verir
            durum_etiketi.config(text="QR Kodu Oluşturuldu ve Kaydedildi.")
```

## 3. Adım: Kullanıcı Arayüzünün (GUI) Kurulması

Bu adımda penceremizi çiziyoruz ve içine giriş kutusu (`Entry`), buton (`Button`) ve metinleri (`Label`) grid (ızgara) sistemiyle yerleştiriyoruz.

**NOT:** Bu hücreyi çalıştırdığınızda bilgisayarınızda bir pencere açılacak ve Notebook bu hücrede asılı (çalışır durumda) kalacaktır. Devam etmek için işiniz bitince o pencereyi kapatın.

```
# Ana pencereyi oluşturur
uygulama_pencere = tk.Tk()
uygulama_pencere.title("QR Kodu Oluşturucusu")

# Gerekli parçaları (widget'ları) yaratıyoruz
etiket = tk.Label(uygulama_pencere, text="URL'yi Girin: ")
url_girdi = tk.Entry(uygulama_pencere, width=40)
qr_kodu_butonu = tk.Button(uygulama_pencere, text="QR Kodu Oluştur", command=qr_kodu_olustur)
durum_etiketi = tk.Label(uygulama_pencere, text="")

# Yaratılan parçaları grid (ızgara) sistemi ile ekrandaki yerlerine oturtuyoruz
etiket.grid(row=0, column=0, padx=10, pady=10) # 1. Satır, 1. Sütun
url_girdi.grid(row=0, column=1, padx=10, pady=10) # 1. Satır, 2. Sütun
qr_kodu_butonu.grid(row=1, column=0, columnspan=2, padx=10, pady=10) # 2. Satır, Ortalanmış
durum_etiketi.grid(row=2, column=0, columnspan=2, padx=10, pady=10) # 3. Satır, Ortalanmış

# Pencerenin kapanana kadar dinleme (bekleme) modunda açık kalmasını sağlar
uygulama_pencere.mainloop()
```
