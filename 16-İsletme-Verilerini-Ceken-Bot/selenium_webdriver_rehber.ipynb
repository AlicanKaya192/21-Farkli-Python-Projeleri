## 1. Selenium Nedir? <a id='bolum1-1'></a>

**Selenium**, web tarayıcılarını programatik olarak kontrol etmek için kullanılan açık kaynaklı bir framework'tür. 2004 yılında Jason Huggins tarafından ThoughtWorks'te geliştirilmiştir.

### Selenium Bileşenleri

| Bileşen | Açıklama | Kullanım |
|---------|----------|----------|
| **Selenium WebDriver** | Tarayıcıyı doğrudan kontrol eder | Otomasyon, scraping |
| **Selenium IDE** | Tarayıcı eklentisi (kayıt/oynat) | Hızlı test kaydı |
| **Selenium Grid** | Paralel test çalıştırma | Dağıtık test ortamları |

### Selenium Ne İçin Kullanılır?

| Kullanım Alanı | Açıklama |
|---------------|----------|
| 🧪 **Fonksiyonel Test** | Web uygulamalarının doğru çalışıp çalışmadığını test etme |
| 🔄 **Regresyon Testi** | Yeni değişikliklerin mevcut işlevselliği bozup bozmadığını kontrol |
| 🕷️ **Web Scraping** | Dinamik web sitelerinden veri çekme |
| 🤖 **Görev Otomasyonu** | Tekrarlayan web görevlerini otomatikleştirme |
| 📊 **Performans İzleme** | Sayfa yükleme sürelerini ölçme |

### Selenium vs Alternatifler

| Özellik | Selenium | Playwright | Puppeteer | Beautiful Soup | Scrapy |
|---------|----------|-----------|-----------|---------------|--------|
| Dil Desteği | Python, Java, C#, JS, Ruby | Python, Java, JS, .NET | Node.js | Python | Python |

## 2. WebDriver Nedir ve Nasıl Çalışır? <a id='bolum1-2'></a>

**WebDriver**, W3C tarafından standartlaştırılmış bir protokoldür. Python kodunuz ile gerçek tarayıcı arasında köprü görevi görür.

### WebDriver Mimarisi

```
┌───────────────────┐                    ┌───────────────────┐                    ┌───────────────────┐
│                   │   HTTP/JSON        │                   │  DevTools          │                   │
│   Python Script   │   (W3C WebDriver   │   Browser Driver  │  Protocol          │   Gerçek Tarayıcı │
│                   │    Protocol)       │                   │  (CDP)             │                   │
│  ┌─────────────┐  │ ──────────────►   │  ┌─────────────┐  │ ──────────────►   │  ┌─────────────┐  │
│  │ selenium    │  │                    │  │ chromedriver│  │                    │  │ Chrome.exe  │  │
│  │ kütüphanesi │  │ ◄──────────────   │  │ msedgedriver│  │ ◄──────────────   │  │ msedge.exe  │  │
│  └─────────────┘  │   JSON Response    │  │ geckodriver │  │  Sonuç             │  │ firefox.exe │  │
│                   │                    │  └─────────────┘  │                    │  └─────────────┘  │
└───────────────────┘                    └───────────────────┘                    └───────────────────┘
     İstemci (Client)                      Sunucu (Server)                          Tarayıcı
```

### İletişim Akışı

1. **Python kodu** → `driver.find_element(By.ID, "search")` çağrısı yapar
2. **Selenium client** → Bu çağrıyı HTTP POST isteğine dönüştürür: `POST /session/{id}/element`
3. **Driver (chromedriver/msedgedriver)** → HTTP isteğini alır, tarayıcının DevTools API'sine iletir
4. **Tarayıcı** → DOM'da elementi bulur ve sonucu döner
5. **Driver** → Sonucu JSON formatında Python'a geri gönderir

### W3C WebDriver Protokolü

| HTTP Metodu | Endpoint | Açıklama |
|-------------|----------|----------|
| POST | `/session` | Yeni oturum başlat |
| DELETE | `/session/{id}` | Oturumu sonlandır (quit) |
| POST | `/session/{id}/url` | URL'ye git |
| POST | `/session/{id}/element` | Element bul |
| POST | `/session/{id}/element/{id}/click` | Elemente tıkla |
| POST | `/session/{id}/element/{id}/value` | Metin yaz |
| GET | `/session/{id}/title` | Sayfa başlığını al |

## 3. Selenium Tarihçesi ve Sürümleri <a id='bolum1-3'></a>

| Yıl | Sürüm | Önemli Değişiklik |
|-----|-------|------------------|
| 2004 | Selenium 1 (RC) | İlk sürüm. JavaScript tabanlı. Proxy sunucu gerekiyordu. |
| 2011 | Selenium 2 (WebDriver) | WebDriver birleştirildi. Doğrudan tarayıcı kontrolü. |
| 2016 | Selenium 3 | W3C WebDriver standardı desteği. GeckoDriver zorunluluğu. |
| 2021 | **Selenium 4** | Tam W3C standardı. Relative locators. CDP desteği. |
| 2023 | Selenium 4.6+ | Otomatik driver yönetimi (Selenium Manager). |

### Selenium 4'ün Getirdiği Yenilikler

- ✅ **Selenium Manager**: chromedriver/msedgedriver otomatik indirilir
- ✅ **Relative Locators**: `locate_with()` ile elementleri komşularına göre bulma
- ✅ **CDP Desteği**: Chrome DevTools Protocol'e doğrudan erişim
- ✅ **Yeni pencere türü**: `driver.switch_to.new_window('tab')` veya `'window'`
- ✅ **Geliştirilmiş waits**: Daha stabil bekleme mekanizmaları

## 17. Custom Expected Conditions <a id='bolum4-17'></a>

````python
# === CUSTOM EXPECTED CONDITIONS ===

print("🛠️ Custom Expected Conditions - Kendi Bekleme Koşulunuzu Yazma")
print("=" * 70)

print("""
# ─── Lambda ile basit koşul ───
wait.until(lambda d: len(d.find_elements(By.CLASS_NAME, "result")) >= 10)
# → Sonuç sayısı 10 veya daha fazla olana kadar bekle

# ─── Fonksiyon ile ───
def element_has_css_class(locator, css_class):
    def check(driver):
        element = driver.find_element(*locator)
        if css_class in element.get_attribute("class"):
            return element
        return False
    return check

# Kullanım:
element = wait.until(element_has_css_class((By.ID, "panel"), "visible"))

# ─── Class ile (en temiz yöntem) ───
class element_count_is:
    def __init__(self, locator, count):
        self.locator = locator
        self.count = count
    
    def __call__(self, driver):
        elements = driver.find_elements(*self.locator)
        if len(elements) >= self.count:
            return elements
        return False

# Kullanım:
items = wait.until(element_count_is((By.CLASS_NAME, "item"), 20))

# ─── Scroll yüksekliği değişikliği bekleme ───
def scroll_height_changed(panel_element, old_height):
    def check(driver):
        new_height = driver.execute_script(
            "return arguments[0].scrollHeight", panel_element
        )
        return new_height if new_height != old_height else False
    return check
""")
````

````python
# === SCROLL İŞLEMLERİ ===

print("🔄 Scroll (Kaydırma) İşlemleri - Tüm Yöntemler")
print("=" * 70)

print("""
# ─── 1. Sayfayı Kaydırma ───

# Sayfanın en altına kaydır
driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")

# Sayfanın en üstüne kaydır
driver.execute_script("window.scrollTo(0, 0)")

# Belirli bir miktar aşağı kaydır (300 piksel)
driver.execute_script("window.scrollBy(0, 300)")

# Yumuşak kaydırma
driver.execute_script("window.scrollTo({top: 1000, behavior: 'smooth'})")

# ─── 2. Elemente Kaydırma ───

# Elementi görünür yap
driver.execute_script("arguments[0].scrollIntoView(true)", element)

# Yumuşak ve merkezde
driver.execute_script(
    "arguments[0].scrollIntoView({behavior: 'smooth', block: 'center'})", 
    element
)

# Selenium 4 native scroll (ActionChains)
from selenium.webdriver.common.action_chains import ActionChains
ActionChains(driver).scroll_to_element(element).perform()

# ─── 3. Belirli Bir Paneli/Divi Kaydırma ───
# (Google Haritalar sonuç paneli gibi overflow:scroll olan elementler)

panel = driver.find_element(By.XPATH, "//div[@role='feed']")

# Paneli en alta kaydır
driver.execute_script(
    "arguments[0].scrollTop = arguments[0].scrollHeight", panel
)

# Paneli 500 piksel aşağı kaydır
driver.execute_script(
    "arguments[0].scrollTop += 500", panel
)

# Panelin mevcut scroll pozisyonunu al
pos = driver.execute_script("return arguments[0].scrollTop", panel)

# ─── 4. Infinite Scroll (Sonsuz Kaydırma) ───

import time

last_height = driver.execute_script("return document.body.scrollHeight")
while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
    time.sleep(2)  # Yüklenmesini bekle
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break  # Yeni içerik yüklenmedi → sona gelindi
    last_height = new_height
""")

print("⚠️ ÖNEMLİ: Google Haritalar gibi sitelerde kaydırma SAYFADA değil,")
print("   sol sidebar PANELİNDE yapılır. window.scrollTo() işe yaramaz!")
print("   Doğru eleman bulunup arguments[0].scrollTop kullanılmalıdır.")
````

## 35. Page Object Model (POM) <a id='bolum9-35'></a>

````python
# === PAGE OBJECT MODEL (POM) ===

print("🏗️ Page Object Model (POM) - Endüstri Standardı Tasarım Deseni")
print("=" * 75)

print("""
POM, her web sayfasını bir Python sınıfı olarak temsil eder.
Element seçicileri ve sayfa işlemleri bu sınıfta kapsüllenir.
Test kodu ve sayfa kodu birbirinden ayrılır.

# ─── Sayfa Sınıfı (Page Object) ───

class LoginPage:
    # Element seçicileri (locator'lar)
    USERNAME_INPUT = (By.ID, "username")
    PASSWORD_INPUT = (By.ID, "password")
    LOGIN_BUTTON = (By.CSS_SELECTOR, "button[type='submit']")
    ERROR_MESSAGE = (By.CLASS_NAME, "error-message")
    
    def __init__(self, driver):
        self.driver = driver
        self.wait = WebDriverWait(driver, 10)
    
    def navigate(self):
        self.driver.get("https://example.com/login")
        return self
    
    def enter_username(self, username):
        field = self.wait.until(EC.visibility_of_element_located(self.USERNAME_INPUT))
        field.clear()
        field.send_keys(username)
        return self  # Zincirleme kullanım için
    
    def enter_password(self, password):
        field = self.wait.until(EC.visibility_of_element_located(self.PASSWORD_INPUT))
        field.clear()
        field.send_keys(password)
        return self
    
    def click_login(self):
        btn = self.wait.until(EC.element_to_be_clickable(self.LOGIN_BUTTON))
        btn.click()
        return DashboardPage(self.driver)  # Sonraki sayfayı döndür
    
    def login(self, username, password):
        \"\"\"Tam login akışı (convenience method)\"\"\" 
        return self.enter_username(username)\
                   .enter_password(password)\
                   .click_login()
    
    def get_error_message(self):
        return self.wait.until(
            EC.visibility_of_element_located(self.ERROR_MESSAGE)
        ).text


# ─── Kullanım ───

driver = webdriver.Edge()
login_page = LoginPage(driver).navigate()
dashboard = login_page.login("admin", "password123")

# Avantajları:
#   ✅ Selector değişirse sadece 1 yerde güncellenir
#   ✅ Test kodu okunabilir ve temiz olur
#   ✅ Kod tekrarı önlenir (DRY prensibi)
#   ✅ Bakımı kolaydır
""")
````
